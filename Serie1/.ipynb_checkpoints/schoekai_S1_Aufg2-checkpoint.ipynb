```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

import scipy.constants as con
```

```python
#w(x,t) = sin(x+ct)
x = np.arange(1,6,0.1)
t = np.arange(1,6,0.05)
x, t = np.meshgrid(x,t)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Axes3D.plot_wireframe(ax,x, t, np.sin(x+t), rstride=10, cstride=10)

plt.show()
```

```python
#v(x,t) = sin(x+ct)+cos(2x+2ct)
x = np.arange(1,6,0.1)
t = np.arange(1,6,0.05)
x, t = np.meshgrid(x,t)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Axes3D.plot_wireframe(ax,x, t, np.sin(x+t)+np.cos(2*x+2*t), rstride=10, cstride=10)

plt.show()
```
