```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

import scipy.constants as con
```

```python
#a)
v = np.arange(0,100, 10 ** 0)
alpha = np.arange(0, np.pi/2, np.pi/2 * 10 ** -2)
#use a meshgird which calculates an area
v, alpha = np.meshgrid(v,alpha)
W = v ** 2 * np.sin(2 * alpha) / con.g

# W maximal bei alpha = 0.758
print(np.arcsin(np.max(W) * con.g / 99 ** 2 ) / 2)

# set up a figure twice as wide as it is tall
fig = plt.figure(figsize=plt.figaspect(0.5))
# plot a 3D surface like in the example mplot3d/surface3d_demo
ax = fig.add_subplot(111, projection='3d')
Axes3D.set_title(ax,"W(v,alpha)")
Axes3D.set_zlabel(ax,'z')
Axes3D.set_ylabel(ax,'y')
Axes3D.set_xlabel(ax,'x')
wire = Axes3D.plot_wireframe(ax,v, alpha, W,rstride=10, cstride=10)

# set up the axes for the second plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Axes3D.plot_surface(ax,v, alpha, W, rstride=10, cstride=10)
plt.show()
plt.savefig('plot.png')
```

```python
#b)
#p(V,T) = R * T / V
V = np.arange(0.01, 0.2 , 1e-2)
T = np.arange(0.1,1e4,1e3)
V, T = np.meshgrid(V,T)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Axes3D.plot_surface(ax,V, T, con.R * T / V, rstride=10, cstride=10)

#V(p,T) = R * T / p
p = np.arange(1e4,1e5,1e4)
T = np.arange(0.1,1e4,1e3)
p, T = np.meshgrid(p,T)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Axes3D.plot_surface(ax,p, T, con.R * T / p, rstride=10, cstride=10)

#T(p,V) = p * V / R
p = np.arange(1e4,1e6,1e3)
V = np.arange(0.1,10,1e-2)
p, V = np.meshgrid(p,V)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Axes3D.plot_surface(ax,p, V, p * V / con.R, rstride=10, cstride=10)

plt.show()
```
